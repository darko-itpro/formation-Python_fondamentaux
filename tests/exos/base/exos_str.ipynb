## Exercice 2

```python
import pyflix.datasource as ds

EPISODE_DURATION = 23
episodes = 7

view_duration = episodes * EPISODE_DURATION

start_hour, start_minutes = ds.get_start_time().split('h')

start_hour = int(start_hour)
start_minutes = int(start_minutes)

start_time = start_hour * 60 + start_minutes
end_time = start_time + view_duration

print('Fin à {}h{:02}'.format(end_time // 60, end_time % 60))
```
