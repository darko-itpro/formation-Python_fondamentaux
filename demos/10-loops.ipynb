# Les boucles

```python
from pkg_resources import parse_version

camelot = ["Arthur", "Merlin", "Lancelot", "Robin", "Perceval", "Karadoc"]
```

## Boucles `For`

La structure est `for element in sequence`

```python
for knight in camelot:
    print(knight.upper())

print(knight)
```

```python
for index, knight in enumerate(camelot):
    if index % 2:
        print(">", knight)
    else:
        print("-", knight)
```

```python
for _ in range(4):
    print("turn left")
```

## Boucles `while`

```python
value = 1
previous = 1

while value < 500:
    print(value / previous)
    value, previous = (value + previous, value)
```

```python
camelot = [("Arthur", "king"), ("Merlin", "wizard"),
           ("Lancelot", "knight"), ("Robin", "knight"),
           ("Perceval", "knight"), ("Karadoc", "knight")]
```

```python
for name, job in camelot:
    print(name)
    if job == "wizard":
        continue
    print("is a mighty knight")
```

```python
for name, job in camelot:
    print(name)
    if job == "wizard":
        print("You have a wizard")
        break
else:
    print("You NEED a wizard")
```

## List Comprehension

```python
camelot = ["Arthur", "Merlin", "Lancelot", "Robin", "Perceval", "Karadoc"]
```

```python
[name.upper() for name in camelot]
```

```python
camelot = [("Arthur", "king"), ("Merlin", "wizard"),
           ("Lancelot", "knight"), ("Robin", "knight"),
           ("Perceval", "knight"), ("Karadoc", "knight")]
```

```python
[peon for peon in camelot if peon[1] != "wizard"]
```

```python
[name.upper()
 for name, job in camelot
 if job != "wizard"]
```
