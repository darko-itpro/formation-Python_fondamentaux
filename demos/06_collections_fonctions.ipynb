## Paramètres optionnels

```python
def say_hello(name="World"):
    print(f'Hello {name}')
```

```python
say_hello()
say_hello('John')
```

```python
def create_account(my_id, value:int=100, overdraft:bool=False):
    print(my_id, value, overdraft)
```

```python
create_account("My id", 1000, True)
```

```python
create_account('My Id', overdraft=True)
```
