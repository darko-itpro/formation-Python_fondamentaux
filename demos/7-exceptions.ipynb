## Capture des exceptions

```python
def convert(value):
    v = int(value)
    return v

user_input = "quarante deux"

try:
    value = convert(user_input)
    print(value, "is cool")
except IndexError:
    print("index de l'exception")
except (ValueError, LookupError):
    print("gestion de l'exception")


print("All good")
```
