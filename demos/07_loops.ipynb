# Les boucles

```python
camelot = ["Arthur", "Merlin", "Lancelot", "Robin", "Perceval", "Karadoc"]
```

## Boucle for

```python
for knight in camelot:
    print(knight.upper())
```

```python
for index, knight in enumerate(camelot):
    if index % 2:
        print(f"- {knight}")
    else:
        print(f"+ {knight}")
```

```python
for _ in range(4):
    print("Turne left")
```

## Boucle while

```python
value = 1
previous = 1

while value < 5000:
    print(value / previous)
    value, previous = value + previous, value
```

## Avec slicing

```python
camelot = ["Arthur", "Merlin", "Lancelot", "Robin", "Perceval", "Karadoc"]
```

```python
start = 0
step = 2

while group := camelot[start : start + step] :
    print(group)
    start = start + step
```

## Interruption de boucles

```python
camelot = [("Arthur", "king"), ("Merlin", "wizard"),
           ("Lancelot", "knight"), ("Robin", "knight"),
           ("Perceval", "knight"), ("Karadoc", "knight")]
```

```python
for name, job in camelot:
    print("Hail", name)
    if job == "wizard":
        continue
    print("He is a mighty knight")
```

```python
for name, job in camelot:
    print("Hail", name)
    if job == "wizard":
        print("you have a wizard")
        break
```
