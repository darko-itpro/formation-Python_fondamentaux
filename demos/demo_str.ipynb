```python
'Arthur'
```

```python
'Arthur'.upper()
```

```python
"Arthur".casefold()
```

```python
"Meaning  \n   of life".split()
```

```python
"Meaning     of life".split(' ')
```

```python
" ".join(['Meaning', 'of', 'life'])
```

```python
"Meaning_of_life".replace('_', ' ')
```

```python
"Prix : {:10.2f} {}".format(42, "€")
```

```python
"Prix : {:.2f} {}".format(42.567, "€")
```

```python
"Price: {1}{0:.2f}".format(76, "$")
```

```python
"Il est {}h{:02}".format(15, 8)
```

```python
"Il est {}h{:02}".format(*divmod(1233, 60))
```

```python
price = 42
currency = "€"
f"Prix : {price:.2f} {currency}"
```

```python
print(f"{price=}")
```
