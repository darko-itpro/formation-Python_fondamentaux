```python
from datetime import datetime
class Training:
    def __init__(self, name:str, duration:int, seats:int=12):
        self.name = name
        self.duration = int(duration)
        self.seats = seats
        self._students = []

    @property
    def students(self):
        return [student for student, _ in self._students]

    def add_student(self, student:str):
        self._students.append((student, datetime.now()))

    def report_students(self):
        print(len(self._students))

    def __str__(self):
        return f"Training {self.name}"

    def __repr__(self):
        return f"Training('{self.name}', {self.duration}, {self.seats})"
```

```python
tp = Training("Python", 5)
td = Training("Django", 4)
#t.set_students(["me", "you"])
```
