## Exercice vitesse

```python
distance:float =19.7
duration:float = 6.892

speed:float = distance / duration
print(f'{speed=}')

print(f"Vitesse de {speed:.2f} m/s")
```

```python
duration = 243

print("La durée est de {}h{:02}".format(*divmod(duration, 60)))
```
