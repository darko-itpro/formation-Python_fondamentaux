```python
camelot = ["Arthur", "Merlin", "Lancelot", "Robin", "Perceval", "Karadoc"]
```

```python
for item in (len(name) for name in camelot):
    print(item)
```

```python
def my_func(kingdom):
    for name in kingdom:
        yield len(name)

    print("Done")
```

```python
for item in my_func(camelot):
    print(item)
```
