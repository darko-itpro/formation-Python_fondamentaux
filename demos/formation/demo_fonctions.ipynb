# Les fonctions

```python
def say_hello():
    print("Hello world")
    print("to all")
```

```python
print(say_hello())
```

```python
def give_hello() -> str:
    value = "Hello given"
    return value
```

```python
print(give_hello())
```

```python
def say_hello_to(name:str):
    print("hello", name)
```

```python
say_hello_to("john")
```

```python
def say_hello(name:str="World"):
    print("hello", name)

say_hello()
say_hello("John")
```

```python
def create_account(id, value:int=100, overdraft:bool=False):
    print(id, value, overdraft)
```

```python
create_account("My id", 1_000_000_000, True)
```

```python
create_account("My id", 1_000_000)
create_account("My Id")
```

```python
create_account(id="my Id", overdraft=True)
```
