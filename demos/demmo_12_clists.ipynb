## Comprehension lists

```python
camelot = ["Arthur", "Merlin", "Lancelot", "Robin", "Perceval", "Karadoc"]
```

```python
[knight.upper() for knight in camelot]
```

```python
[len(knight) for knight in camelot]
```

```python
camelot = [("Arthur", "king"), ("Merlin", "wizard"),
           ("Lancelot", "knight"), ("Robin", "knight"),
           ("Perceval", "knight"), ("Karadoc", "knight")]
```

```python
[name for name, job in camelot if job != "wizard"]
```

```python
camelot = ["Arthur", "Merlin", "Lancelot", "Robin", "Perceval", "Karadoc"]
```

```python
for i in (len(knight) for knight in camelot):
    print(i)
```

```python
def upperize(knights):
    for knight in knights:
        yield knight.upper()
        
    print("after")
    
```

```python
for i in upperize(camelot):
    print("-", i)
```
