```python
kingdom = []

def add_knight(name):
    kingdom.append(name)

add_knight("Lancelot")

print(kingdom)
```

```python
camelot = []

def add_knight(name, kingdom):
    kingdom.append(name)

add_knight("Lancelot", camelot)

print(camelot)
```

```python
camelot = []

def add_knight(name, kingdom):
    kingdom = kingdom.copy()
    kingdom.append(name)

    return kingdom

print(add_knight("Lancelot", camelot))

print(camelot)
```
