```python
def func(value):
    if value < 10:
        raise ValueError("value lower than 10")
    return value * 2
```

```python
func(0)
```

```python
try:
    value = func(5)
except IndexError:
    value = 10
except (ValueError, LookupError):
    value = 0

print(value)
```
