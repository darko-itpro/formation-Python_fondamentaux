```python
value = "toto"

def do_something(data):
    data = int(data)
    return data + 2

try:
    result = do_something(value)
    print("Something done")
except (ValueError, LookupError):
    print("Exception gérée")
    result = None
except KeyError as exc:
    if exc.args[0] != "Le bon message":
        raise exc
    print("gérée autrement")

print("got", result)
print("done")
```
