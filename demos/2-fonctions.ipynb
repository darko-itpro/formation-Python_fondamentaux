# Les fonctions

```python
def say_hello():
    print("Hello World")
    print("To all")
```

```python
print(say_hello())
```

```python
def give_hello() -> str:
    value = "Hello Given"
    return value
```

```python
print(give_hello())
```

```python
def say_hello_to(name: str):
    print("Hello", name)
```

```python
say_hello_to("Me")
```

## Paramètres optionnels

```python
def say_hello(name:str = "World"):
    print("Hello", name)
```

```python
say_hello()
```

```python
say_hello("Me")
```

```python
def create_account(myid: str, value: float = 100, overdraft:bool = False):
    print(myid, value, overdraft)
```

```python
create_account("My Id", 10_000_000_000, True)
```

```python
create_account("My Id", 10_000_000_000)
```

```python
create_account("My Id")
```

```python
create_account("My Id", overdraft=True)
```

```python
def create_account(myid: str, value: float = 100, *, overdraft:bool = False):
    print(myid, value, overdraft)
```

```python
create_account("toto", 10000, False)
```
