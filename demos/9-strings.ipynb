## Formatage de chaines

```python
"Prix : {:.2f} {}".format(42, "€")
```

```python
"Prix : {:.2f} {}".format(42.988, "€")
```

```python
"Price: {1}{0:.2f}".format(53, "$")
```

```python
"Il est {}h{:02}".format(2, 5)
```

```python
price = 42
currency = '€'
f"Prix : {price:.2f} {currency}"
```

```python
value = 42

print(f"{value=}")
```
