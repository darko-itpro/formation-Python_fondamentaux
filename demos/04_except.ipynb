```python
user_value = "quarante deux"
```

```python
def convert(value: int):
    result = int(value) + 2
    return result

print("begin")

try:
    answer = convert(user_value)
    print(answer)

except IndexError:
    print("autre chose")
except (ValueError, LookupError):
    print("gérée")


print("after")
```
