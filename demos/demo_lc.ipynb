# Les Lists Comprehension

```python
camelot = [("Arthur", "king"), ("Merlin", "wizard"),
           ("Lancelot", "knight"), ("Robin", "knight"),
           ("Perceval", "knight"), ("Karadoc", "knight")]
```

```python
[name.upper() for name, _ in camelot]
```

```python
[name.upper()
 for name, job in camelot
 if job =="knight"]
```

# Les générateurs

```python
(name.upper() for name, _ in camelot)
```

```python
for item in (name.upper() for name, _ in camelot):
    print(item)
```

```python
camelot = [("Arthur", "king"), ("Merlin", "wizard"),
           ("Lancelot", "knight"), ("Robin", "knight"),
           ("Perceval", "knight"), ("Karadoc", "knight")]
```

```python
def parcours(kingdom):
    for name, _ in kingdom:
        yield name.upper()

    print("done")
```

```python
for item in parcours(camelot):
    print(item)
```
