# List Comprehension

```python
camelot = [("Arthur", "king"), ("Merlin", "wizard"),
           ("Lancelot", "knight"), ("Robin", "knight"),
           ("Perceval", "knight"), ("Karadoc", "knight")]
```

```python
[peon[0] for peon in camelot]
```

```python
[peon for peon in camelot if peon[1] == "knight"]
```

```python
for_quest = [name
             for name, job in camelot
             if job == "knight"]
```

```python
camelot = [("Arthur", "king"), ("Merlin", "wizard"),
           ("Lancelot", "knight"), ("Robin", "knight"),
           ("Perceval", "knight"), ("Karadoc", "knight")]

(peon[0].upper() for peon in camelot)
```

```python
for item in (peon[0].upper() for peon in camelot):
    print(item)
```

```python
def upperize(kingdom):
    for peon in kingdom:
        yield peon[0].upper()

    print("done")
```

```python
for item in upperize(camelot):
    print(item)
```
