```python
class Training:
    def __init__(self, name:str, duration:int, max_seats:int=12) -> None:
        self.subject = name
        self.duration = duration
        self.max_seats = max_seats
        self.students = []
    
    def add_student(self, name):
        self.students.append(name)
        
    def display_students(self):
        for student in self.students:
            print("-", student)
    
```

```python
p = Training('Python', 5)
```

```python
p.add_student('Elliot')
p.add_student('Kevin Mitnik')
```

```python
p.display_students()
```

```python
p.display_students
```

```python
Training.display_students(p)
```
