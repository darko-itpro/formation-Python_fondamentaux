```python
class Training:
    def __init__(self, name:str, duration:int, max_seats:int=12) -> None:
        self.subject = name
        self.duration = duration
        self.max_seats = max_seats
        self.students = []
        
    def __str__(self) -> str:
        return f"Training {self.subject}"
    
    def __repr__(self) -> str:
        return f"Training('{self.subject}', {self.duration}, {self.max_seats})"
    
    def add_student(self, name):
        self.students.append(name)
        
    def display_students(self):
        for student in self.students:
            print("-", student)
    
```

```python
p = Training('Python', 5)
print(p)
```

```python
print(eval(p.__repr__()))
```

```python
p.add_student('Elliot')
p.add_student('Kevin Mitnik')
```

```python
p.display_students()
```

# Les Dataclass

```python
class Episode:
    def __init__(self, title: str, number: int, season_number: int,
                 duration: int = None, year: int = None):
        self.title = title
        self.number = number
        self.season_number = season_number
        self.duration = duration
        self.year = year

    def __eq__(self, other):
        if not isinstance(other, self.__class__):
            return False

        return (self.number, self.season_number) == (other.number, other.season_number)

    def __str__(self):
        return f'Episode {self.title} s{self.season_number:02}e{self.number:02}'
```

```python
from dataclasses import dataclass, field

@dataclass(frozen=True)
class Episode:
    title:str = field(compare=False)
    number:int
    season_number:int
    duration: int|None = field(default=None, compare=False)
    year: int|None =  field(default=None, compare=False)
    
    
```
