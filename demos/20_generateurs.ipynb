# Les générateurs

```python
camelot = ["Arthur", "Merlin", "Lancelot", "Robin", "Perceval", "Karadoc"]
```

```python
my_gen = (name.upper() for name in camelot)
```

```python
for item in my_gen:
    print(item)
```

```python
def upperize(kingdome):
    for name in kingdome:
        yield name.upper()

    print("done")

for item in upperize(camelot):
    print(item)
```

```python
def fibo(max_value=50):
    value = 1
    previous = 1
    while value < max_value:
        yield value
        value, previous = value + previous, value
```

```python
for i in fibo():
    print(i)
```
