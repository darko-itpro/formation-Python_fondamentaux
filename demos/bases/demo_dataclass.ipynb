# Les Dataclasses

```python
from dataclasses import dataclass, field

@dataclass(frozen=False)
class Episode:
    title: str = field(compare=False)
    season_number: int
    number: int
    duration: int = field(default=None, compare=False)
```

```python
ep1 = Episode("Title", 2, 1, 54)
```

```python
ep1["title"] = "autre tutre"
```

```python
ep1
```

```python
ep2 = Episode('Titre 2', 2, 3)
```

```python
ep2
```

```python
ep3 = Episode("Title", 2, 1, 54)
```

```python
ep1 == ep3
```

```python
help(field)
```

## Bonus : Dicts (et slots)

```python
class Point:
    __slots__ = ('x', 'y')

    def __init__(self, x, y=0):
        self.x = x
        self.y = y

p = Point(6, 8)
```

```python
dir(p)
```

```python
p.__slots__
```

```python
p.z = 0
```

```python
p.x
```

```python
p.__dict__['z'] = 0
```

```python
p.z
```

```python
del p.__dict__
```
