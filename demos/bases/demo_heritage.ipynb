# Heritage et abstraction

```python
from abc import ABC, abstractmethod

class vehicle(ABC):
    def __init__(self, max_speed,):
        self._speed = 0
        self.max_speed = max_speed

    @abstractmethod
    def accelerate(self):
        self._speed += 1

class boat(vehicle):
    pass

class plane(vehicle):
    def __init__(self, max_speed, max_ceiling):
        vehicle.__init__(self, max_speed)
        self.max_ceiling = max_ceiling
        self.height = 0

    def accelerate(self):
        super().accelerate()
        self.height += 1

class Hydroplane(boat, plane):
    pass
```
