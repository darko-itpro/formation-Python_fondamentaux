```python
class Venue:
    def __init__(self, address, room, capacity):
        self.address = address
        self.room = room
        self.capacity = capacity

class Training:
    def __init__(self, subject, duration, venue:Venue, seats=12):
        self._max_seats = seats
        self.duration = duration
        self.name = subject
        self.students = []
        self.venue = venue

    def __len__(self):
        return len(self.students)

    def __contains__(self, item):
        return item in self.students

    def __getitem__(self, item):
        print(item, type(item))

    @property
    def seats(self):
        return min(self._max_seats, self.venue.capacity)

    @property
    def seats_left(self):
        return self.seats - len(self.students)

    def __str__(self):
        return f"Training {self.name}"

    def __repr__(self):
        return f"Training('{self.name}', {self.duration}, {self.seats})"
```

```python
class Training:
    def __init__(self, subject, duration, seats=12):
        self.name = subject

    def give_name(self, name):
        self.name = name.capitalize()
    def show_name(self):
        print(self.name)
```

```python
t = Training("PHP", 12)
```

```python
t.show_name()
```

## Properties en vrai

```python
def get_value(self):
    return self._value

class DemoProp:
    def __init__(self):
        self._value = 10

    def _set_value(self, value):
        self._value = value

    value = property(get_value, _set_value)
```

```python
d = DemoProp()
d._value
```

```python
d.value = 20
```

```python
d.value
```
