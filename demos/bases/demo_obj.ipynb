```python
class Training:
    def __init__(self, subject, duration, seats=12):
        self.seats = seats
        self.duration = duration
        self.name = subject

    def give_name(self, name):
        self.name = name.capitalize()
    def show_name(self):
        print(self.name)

    def __str__(self):
        return f"Training {self.name}"

    def __repr__(self):
        return f"Training('{self.name}', {self.duration}, {self.seats})"
```

```python
t = Training("PHP", 5)
```

```python
eval("Training('PHP', 5, 12)")
```

```python
t.give_name("Python")
```

```python
t.show_name()
```

```python
t.name
```
