# Boucles

```python
camelot = ["Arthur", "Merlin", "Lancelot", "Robin", "Perceval", "Karadoc"]
```

## Boucle for

```python
for knight in camelot:
    print(knight)
```

```python
for index, knight in enumerate(camelot):
    print(knight, index)
```

```python
for _ in range(4):
    print("turn left"
          "")
```

## Boucle while

```python
value = 1
previous = 1

while value < 5000:
    print(value / previous)
    value, previous = value + previous, value

```

## Conditions d'arrêt

```python
camelot = [("Arthur", "king"), ("Merlin", "wizard"),
           ("Lancelot", "knight"), ("Robin", "knight"),
           ("Perceval", "knight"), ("Karadoc", "knight")]

```

```python
for name, job in camelot:
    print("Hail ", name)
    if job == "wizard":
        continue
    print(" he is a mighty knight")
```

```python
for name, job in camelot:
    print(name)
    if job == "wizad":
        print("you have a wizard")
        break
else:
    print("You need a wizard")
```

## Lists Comprehension

```python
camelot = ["Arthur", "Merlin", "Lancelot", "Robin", "Perceval", "Karadoc"]
```

```python
[len(name) for name in camelot]
```

```python
camelot = [("Arthur", "king"), ("Merlin", "wizard"),
           ("Lancelot", "knight"), ("Robin", "knight"),
           ("Perceval", "knight"), ("Karadoc", "knight")]
```

```python
[knight for knight in camelot if knight[1] != 'wizard']
```

```python
[len(name)
 for name, job in camelot
 if job != "wizard"]
```
