# Context Managers

```python
class DemoContext:
    def __init__(self, name):
        self.name = name

    def say_hello(self):
        print(f'Hello from {self.name}')

    def __enter__(self):
        print('enter')

        return self

    def __exit__(self, exc_type, exc_val, exc_tb):
        print("Sortie 1")
        print(f"{exc_type=}")
        print(f"{exc_val=}")
        print(f"{exc_tb=}")
        print("Sortie 2")

        return True
```

```python
print("Avant")
with DemoContext("demo 1") as my_cm1:
    print("dedans")
    my_cm1.say_hello()

print("Après")
```

```python
my_cm2 = DemoContext("demo 2")
with my_cm2:
    print("dedans")
    my_cm2.say_hello()
```

```python
with DemoContext("demo 3"):
    raise ValueError("ValueError")
```
