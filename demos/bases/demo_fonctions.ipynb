# Les fonctions

```python
def say_hello():
    print("Hello world!")
    print("to all")
```

```python
say_hello()
```

```python
def give_hello() -> str:
    value = "Given hello"
    return value
```

```python
print(give_hello())
```

```python
print(say_hello())
```

```python
def say_hello_to(name:str):
    print("Hello", name)
```

```python
say_hello_to("Me")
```

```python
def say_hello(name="World"):
    print("Hello", name)

say_hello()
say_hello("me")
```

```python
def create_account(nid, value: float = 100, overdraft: bool = False):
    print(nid, value, overdraft)

create_account("My id", 120_000_000, True)
create_account("My id", value=100000)
create_account("My id")
create_account("My id", overdraft=True)
```
