## Variadics

```python
def func(*args):
    print(args)
```

```python
func()
```

```python
func(1)
```

```python
func(1, 2)
```

```python
def func(*args, **kwargs):
    print(args, kwargs)
```

```python
func()
```

```python
func(param=1)
```
