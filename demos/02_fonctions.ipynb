# Les fonctions

```python
def say_hello():
    print("Hello World")
    print("to all")
```

```python
print(say_hello())
```

## Retour de fonction

```python
def give_hello() -> str:
    value = "Hello given"
    return value
```

```python
print(give_hello())
```

## Paramètres

```python
def say_hello_to(name:str):
    print("Hello", name)
```

```python
say_hello_to("John")
```
