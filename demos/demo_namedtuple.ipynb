# NamedTuple

```python
from collections import namedtuple
```

```python
Point = namedtuple("Point", ["x", "y", "z", "color"], defaults=(0, [0, 0, 0]))
```

```python
p = Point(6, 18)
```

```python
type(p)
```

```python
p.x
```

```python
p.y
```

```python
p.z
```

```python
p.color
```
