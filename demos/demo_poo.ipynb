```python
class Training:
    
    def __init__(self, subject, duration, seats=12):
        self.subject = subject
        self.duration = duration
        self.max_seats = seats
        self.students = []

    def add_students(self, name):
        self.students.append(name)
    
    def reset_students(self):
        self.students.clear()
    
    def give_students(self):
        print(self.students)
```
