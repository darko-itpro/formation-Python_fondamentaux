# Les fonctions

```python
def say_hello():
    print("Hello World")
    print("To all")

say_hello()
```

```python
say_hello()
```

```python
def give_hello():
    value = "Hello given"
    return value
```

```python
print(give_hello())
```

```python
def say_hello_to(name:str):
    print("Hello", name)

say_hello_to("John")
```

```python
def say_hello(name="World"):
    print("Hello", name)

say_hello()
say_hello("John")
```

```python
def create_account(id:str, value:int = 100, overdraft=False):
    print(id, value, overdraft)
```

```python
create_account("My id", 100_000_000, True)
```

```python
create_account("My id", 100_000_000)
create_account("My id")
```

```python
create_account("My id", overdraft=True)
```
