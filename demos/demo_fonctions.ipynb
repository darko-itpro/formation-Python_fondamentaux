```python
def say_hello():
    print("Hello World!")
    print("To all")
```

```python
print(say_hello())
```

```python
def give_hello():
    value = "Hello given"
    return value
```

```python
returned_value = give_hello()
print(returned_value)
```

```python
def say_hello_to(name:str):
    print("Hello", name)
```

```python
say_hello_to("Me")
```

```python
def say_hello(name:str = "World"):
    print("Hello", name)
```

```python
say_hello("Me")
```
