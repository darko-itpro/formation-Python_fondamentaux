# Exceptions

```python
value = "quarante deux"

def transform(data:int):
    d = int(data) + 2
    return d

print("start")
try:
    v = transform(value)
except IndexError:
    print('géré autrement')
    v = 10
except (ValueError, LookupError):
    print('on a géré')
    v = 0

print("on a eu ", v)
print("end")
```
